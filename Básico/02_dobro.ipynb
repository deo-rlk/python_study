### O Exemplo abaixo demonstra que sem delimitar o campo numero como numérico ele vai dobrar como se fosse uma string(letra)

```python
numero = input("Entre com um número para encontrar o seu dobro: ")

dobro = numero * 2

print("O dobro de ", numero, "é", dobro)
```

#### O Exemplo abaixo demonstra que ao delimitar o campo numero como numérico ele vai dobrar como se fosse um número

```python
numero = input("Entre com um número para encontrar o seu dobro: ")

numero = int(numero) #Não permite mais utilizar a função input como string

dobro = numero * 2

print("O dobro de ", numero, "é", dobro)
```

```python
numero = int(input("Entre com um número inteiro para calcular a sua raiz quadrada: "))

raiz = numero ** (1/2)
raiz = round(raiz, 2)
print(f"Raiz quadrada de {numero} é {raiz}")
```
