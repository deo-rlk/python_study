#### Existem diversas maneiras de usar Boolean dentro de código:
- Operações de comparação ( == , != , > , < , >= , <= )
- Estruturas condicionais (IF, ELIF, ELSE)
- Operadores lógicos (and , or , not)

```python
# Testando valores
print(10>9)
print(10 == 9)
print(10 < 9)
```

```python
print(bool("Hello"))
print(bool(""))
print(bool(15))
print(bool(0))
```

```python
numbers = [1,2,3,4,5]

if (count := len(numbers)) > 3:
    print(f"List has {count} elements")
```
