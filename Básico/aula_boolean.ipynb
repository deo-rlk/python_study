```python
idade = int(input("Insira a sua idade:"))

if idade >= 70:
    print("Cuidado com a bebida.")
    print("Consulte o seu médico.")

elif idade >= 18:
    print("Você pode comprar cerveja!")
    print("Beba com moderação!")

elif idade == 17:
    print("Você ainda não pode comprar bebida nem cerveja.")
    print("Quer uma coca cola?")

else:
    print("Você não pode comprar bebida!")
    print("Vai querer algo?")
```

```python
texto = """
Escolha a sua água para comprar
(1) Água mineral com gás
(2) Água mineral natural"""


opcao = input(texto)

if opcao == '1':
    print("Sua conta deu: R$1,50")
elif opcao == '2':
    print("Sua conta deu: R$2,50")
else:
    print("Entre com uma opção válida!")
```

```python
texto = """
Escolha a sua água para comprar
(1) Água mineral com gás
(2) Água mineral natural"""


opcao = input(texto)

conta = 0
if opcao == '1':
    conta = 1.50

elif opcao == '2':
    conta = 2.50

if conta == 0:
    print("Entre com uma opção válida") 

else:
    print(f"Sua conta é R${conta}")
```
